```python
duplicates = df.duplicated()
print(duplicates.any())  
```

```python
def cap_outliers(df, columns):
    df_capped = df.copy()
    for col in columns:
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        mean_value = df_capped[col].mean()
        df_capped[col] = df_capped[col].apply(lambda x: mean_value if x < lower_bound or x > upper_bound else x)
    return df_capped
df=cap_outliers(df,['study_hours_per_week','sleep_hours_per_day','attendance_percentage'])
```

```python
X=df.drop('result',axis=1)
y=df['result']
```

```python
from imblearn.over_sampling import RandomOverSampler
smote = RandomOverSampler()
X_res, y_res = smote.fit_resample(X, y)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,auc,roc_curve,confusion_matrix
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,random_state=42,test_size=0.2)
```

```python
random=RandomForestClassifier(class_weight='balanced')
```

```python
random.fit(X_train,y_train)
y_pred_test=random.predict(X_test)
y_pred_train=random.predict(X_train)
```

```python
#Train
```

```python
print('accuracy_score Train:',accuracy_score(y_pred_train,y_train))
print('accuracy_score Test:',accuracy_score(y_pred_test,y_test))
```

```python
print('classification_report Train:\n',classification_report(y_train,y_pred_train))
print('classification_report Test:\n',classification_report(y_test,y_pred_test))
```

```python
print("Train")
confu=confusion_matrix(y_train,y_pred_train)
sns.heatmap(confu,annot=True,fmt='.2f')
plt.show()
print('TEST')
confu=confusion_matrix(y_test,y_pred_test)
sns.heatmap(confu,annot=True,fmt='.2f')
plt.show()
```

# NB

```python
from sklearn.preprocessing import KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
X_numeric_binned = kbins.fit_transform(X_res[cols])
```

```python
X_transformed=X_res
X_transformed[cols]=X_numeric_binned
```

```python
from sklearn.model_selection import train_test_split
X_train_nb, X_test_nb, y_train_nb, y_test_nb = train_test_split(X_transformed, y_res, test_size=0.2, random_state=42, stratify=y_res)
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import classification_report, roc_auc_score
nb = CategoricalNB()
nb.fit(X_train_nb, y_train_nb)
y_pred_test_nb = nb.predict(X_test_nb)
y_pred_train_nb=nb.predict(X_train_nb)
print('Train')
print('Accuracy:',accuracy_score(y_pred_train_nb,y_train_nb))
print('Classification:\n',classification_report(y_pred_train_nb,y_train_nb))
print('Test')
print('Accuracy:',accuracy_score(y_pred_test_nb,y_test_nb))
print(classification_report(y_test_nb, y_pred_test_nb))
```

```python
print("Train")
confu=confusion_matrix(y_train_nb,y_pred_train_nb)
sns.heatmap(confu,annot=True,fmt='.2f')
plt.show()
print('TEST')
confu=confusion_matrix(y_test_nb,y_pred_test_nb)
sns.heatmap(confu,annot=True,fmt='.2f')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
# Random Forest
y_prob_rf = random.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
# Naive Bayes
y_prob_nb = nb.predict_proba(X_test_nb)[:, 1]
fpr_nb, tpr_nb, _ = roc_curve(y_test_nb, y_prob_nb)
roc_auc_nb = auc(fpr_nb, tpr_nb)
plt.plot(fpr_nb, tpr_nb, label=f'Naive Bayes (AUC = {roc_auc_nb:.2f})')
# Random guess line
plt.plot([0, 1], [0, 1], 'r--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Two Models')
plt.legend()
plt.show()
```
